```python
import numpy as np
import time
import matplotlib.pyplot as plt

def multiply_matrices(A, B):
    """
    Multiplica dos matrices A y B y retorna el producto junto al tiempo de ejecución.
    
    Parameters:
        A (np.ndarray): Primera matriz.
        B (np.ndarray): Segunda matriz.
    
    Returns:
        tuple: Producto de A y B, y tiempo de ejecución en segundos.
    """
    start_time = time.time()
    C = np.dot(A, B)
    end_time = time.time()
    execution_time = end_time - start_time
    return C, execution_time

def generate_random_matrix(n):
    """
    Genera una matriz cuadrada aleatoria de tamaño n x n.
    
    Parameters:
        n (int): Tamaño de la matriz.
    
    Returns:
        np.ndarray: Matriz generada aleatoriamente.
    """
    return np.random.rand(n, n)

def main():
    # Generar 500 puntos en el intervalo [2, 10000].
    # Nota: Para tamaños muy grandes se requiere mucha memoria y tiempo de cómputo.
    # Es recomendable ajustar el rango o número de puntos en función de la capacidad de su máquina.
    sizes = np.linspace(2, 10000, num=500, dtype=int)
    times = []

    for n in sizes:
        # Generar matrices aleatorias
        A = generate_random_matrix(n)
        B = generate_random_matrix(n)
        # Multiplicar matrices y medir el tiempo
        _, t = multiply_matrices(A, B)
        times.append(t)
        print(f"Tamaño: {n}x{n} - Tiempo: {t:.6f} segundos")

    # Graficar los tiempos de ejecución
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, times, marker='o', markersize=3, linestyle='-')
    plt.xlabel('Tamaño de la matriz (n)')
    plt.ylabel('Tiempo de ejecución (segundos)')
    plt.title('Tiempo de ejecución para la multiplicación de matrices')
    plt.grid(True)
    plt.savefig('multiplicacion_matrices.png')
    plt.show()

if __name__ == "__main__":
    main()
```
